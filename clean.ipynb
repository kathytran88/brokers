```python
import pandas as pd
```

```python
df = pd.read_excel('ibba_brokers.xlsx')
df.head()
```

```python
new = ''
for phone in df['phone']:
    phone = str(phone)
    for character in phone:
        if character.isnumeric() == True:
            new += character 
    new += ' '
print(new)
```

```python
new_phone = new.split(' ')
print(new_phone)
```

```python
print(len(new_phone))
```

```python
del new_phone[2582]
```

```python
del df['phone']
df.head()
```

```python
df['phone number'] = new_phone
df.head()
```

```python
state = []
city = []
for place in df['location']:
    place2 = place.split(',')
    city.append(place2[0])
    state.append(place2[len(place2) - 1])

print(state)
```

```python
df['State'] = state 
df['City'] = city
df.head()
```

```python
df = df.sort_values(by='State')
df.head()
```

```python
del df['index']
```

```python
firstname = []
lastname = []
for name in df['name']:
    name2 = name.split(' ')
    if len(name2) == 2:
        firstname.append(name2[0])
        lastname.append(name2[1])
    elif len(name2) > 2:
        first = name2[:-2]
        firstname.append(first)
        last = name2[-1]
        lastname.append(last)
        
print(firstname)
print(lastname)
```

```python
print(len(firstname))
print(len(lastname))
```

```python
df['First Name'] = firstname 
df['Last Name'] = lastname 
df.head()
```

```python
phone2x = []

for phone in df['phone number']:
    if len(phone) > 10 and phone[0] == '1':
        new = phone[1:]
        phone2x.append(new)
    else:
        phone2x.append(phone)
print(phone2x)
```

```python
df['phone'] = phone2x
```

```python
invalid_phones = df[(df['phone'].apply(len) != 10) & (df['phone'].apply(len) != 0)]
print(invalid_phones['name'].tolist())
```

```python
df.rename(columns = {'name': 'Full Name' , 'company': 'Company', 'location': 'Location', 'profile': 'Profile Link', 'email': 'Email', 'phone': 'Corporate Phone'}, inplace=True)
df.head()
```

```python
for c in range(len(df['Corporate Phone'])):
    if len(df['Corporate Phone'][c]) > 0:
        df['Corporate Phone'][c] = '+1 ' + df['Corporate Phone'][c][:3] + '-' + df['Corporate Phone'][c][3:6] + '-' + df['Corporate Phone'][c][6:]

df.head()
```

```python
df = df[['Company', 'First Name', 'Last Name', 'Full Name', 'Email', 'Corporate Phone','Location', 'City', 'State', 'Profile Link']]
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.to_excel('broker_new.xlsx')
```

```python
# Step 2: Export to Excel with clickable links
excel_writer = pd.ExcelWriter('official.xlsx', engine='xlsxwriter')
df.to_excel(excel_writer, index=False, sheet_name='Brokers')

# Get the xlsxwriter workbook and worksheet objects
workbook = excel_writer.book
worksheet = excel_writer.sheets['Brokers']

# Write the links to the Excel file, skipping NaN values
for row_num, link in enumerate(df['Profile Link'], start=1):
    if pd.notna(link):
        worksheet.write_url(row_num, df.columns.get_loc('Profile Link'), link, string=link)

# Save the Excel file
excel_writer._save()
```
